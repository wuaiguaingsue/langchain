# 摘要检查链
本笔记本展示了一些使用 LLMSummarizationCheckerChain 处理不同类型文本的示例。它与 `LLMCheckerChain` 有一些明显的区别，因为它对输入文本（或摘要）的格式没有任何假设。
此外，由于 LLM 在事实检查时容易产生幻觉或被上下文混淆，多次运行检查器有时会更有益。它通过将重写的"真实"结果反馈给自身，并检查"事实"的真实性来实现这一点。从下面的例子可以看出，这种方法在得到一个普遍真实的文本方面非常有效。

您可以通过设置 `max_checks` 参数来控制检查器运行的次数。默认值为 2，但如果您不想进行二次检查，可以将其设置为 1。

```python
from langchain.chains import LLMSummarizationCheckerChain
from langchain_openai import OpenAI

llm = OpenAI(temperature=0)
checker_chain = LLMSummarizationCheckerChain.from_llm(llm, verbose=True, max_checks=2)
text = """
以下是詹姆斯·韦伯太空望远镜（JWST）最近的一些发现，您 9 岁的孩子可能会感兴趣：
• 2023年，JWST发现了一些被称为"绿豆"的星系。它们之所以得到这个名字，是因为它们像豆子一样小、圆而且呈绿色。
• 望远镜捕捉到了超过130亿年前的星系图像。这意味着这些星系的光芒花了超过130亿年才到达我们这里。
• JWST拍摄到了第一张太阳系外行星的照片。这些遥远的世界被称为"系外行星"。Exo的意思是"来自外部"。
这些发现可以激发孩子对宇宙无限奇迹的想象力。"""
checker_chain.run(text)
```

```python
from langchain.chains import LLMSummarizationCheckerChain
from langchain_openai import OpenAI

llm = OpenAI(temperature=0)
checker_chain = LLMSummarizationCheckerChain.from_llm(llm, verbose=True, max_checks=3)
text = "格陵兰海是北冰洋的一个外围部分，位于冰岛、挪威、斯瓦尔巴群岛和格陵兰之间。它的面积为465,000平方英里，是世界五大洋之一，与太平洋、大西洋、印度洋和南大洋并列。它是五大洋中最小的，几乎完全被水覆盖，其中一部分以冰川和冰山的形式冻结。这片海域以格陵兰岛命名，是北冰洋通向大西洋的主要出口。它经常结冰，因此航行受限，被认为是挪威海的北部分支。"
checker_chain.run(text)
```
